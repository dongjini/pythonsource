## matplotlib (https://matplotlib.org/stable/index)

- 다양한 형태의 그래프를 통해서 데이터 시각화를 할 수 있는 라이브러리

```python
# 라이브러리 로드
import matplotlib.pyplot as plt
import koreanize_matplotlib 
```

```python
x= [1,2,3]
y= [2,4,8]
```

#### 1. 선 굵기 : linewidth

```python
g = plt.plot(x, y, linewidth=100)
```

#### 2. 마커
- https://matplotlib.org/stable/gallery/lines_bars_and_markers/marker_reference.html


```python
g = plt.plot(x, y, marker='o')
```

```python
# 마커 사이즈 변경

g = plt.plot(x, y, marker='o', markersize=10)
```

```python
g = plt.plot([4,5,6], marker='H')
g = plt.plot([3,4,5], marker='d')
g = plt.plot([2,3,4], marker='x')
g = plt.plot([1,2,3], marker=11)
g = plt.plot([0,1,2], marker="$\u266B$", markersize=10)

```

```python
# markeredgecolor : 마커 테두리 색상
# markerfacecolor : 마커 색상


g = plt.plot([0,1,2], marker="$\u266B$", markersize=20,
              markeredgecolor='red', markerfacecolor='black')
```

#### 3. 선 모양 : linestyle(https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html)

```python
g = plt.plot(x,y, linestyle=":", linewidth=5)
```

```python
g = plt.plot(x,y, linestyle="--", linewidth=5)
```

```python
g = plt.plot(x,y, linestyle="-.", linewidth=5)
```

#### 4. 선 색상 : color

- https://matplotlib.org/stable/api/colors_api.html#module-matplotlib.colors

```python
# r : red
# g : greed
# b : blue

g = plt.plot(x,y,color='r')
```

```python
g = plt.plot(x,y,color='b')
```

```python
g = plt.plot(x,y,color='g')
```

#### 5. 포맷 - fmt = '[marker][line][color]'

```python
g = plt.plot(x,y, 'o')
```

```python
g = plt.plot(x,y, 'ro')
```

```python
g = plt.plot(x,y, 'ro--')
```

```python
g = plt.plot(x,y, 'bv:')
```

```python
g = plt.plot(x,y, 'go')
```

#### 6. 축약(https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)

```python
# mfc : markerfacecolor
# mec : markeredgecolor
# ls : linestyle
# ms : marker size

g = plt.plot(x,y, marker='o', mec='b', mfc='r', ls=':', ms=10)
```

#### 7. 투명도

```python
g = plt.plot(x,y, marker='o', mec='b', mfc='r', ls='-', ms=10, alpha=0.3)
```

#### 8. 그래프 크기

```python

plt.figure(figsize=(20,3))
g = plt.plot(x,y, marker='o', mec='b', mfc='r', ls='-', ms=10, alpha=0.3)
```

```python
plt.figure(figsize=(3,6))
g = plt.plot(x,y)
```

```python
# dpi : 확대
plt.figure(figsize=(10,5), dpi=300)
g = plt.plot(x,y)
```

#### 9. 배경색

```python
plt.figure(figsize=(10,5), facecolor='pink')
g = plt.plot(x,y)
```
