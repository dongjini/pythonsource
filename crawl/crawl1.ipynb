### 크롤링 vs 스크랩핑
- 크롤링 : 주기적으로, 무작위로 데이터 수집하는 것(데이터의 양을 우선으로 수집하는 방식)
    - 사이트에서 혀용하는 경로만 가능
- 스트랩핑 : 어떤특정 카테고리를 선택해서 데이터를 수집(크롤링보다 작으 개념)

```python
import urllib.request as req

url = "http://google.com"

# urlretrieve
# 요청하는 url 정보를 로컬파일로 저장

file1, header1 = req.urlretrieve(url, "./google.html")
print(header1)
```

```python
url = "https://upload.wikimedia.org/wikipedia/commons/thumb/c/c3/Python-logo-notext.svg/1200px-Python-logo-notext.svg.png"

# urlretrieve
# 요청하는 url 정보를 로컬파일로 저장

file1, header1 = req.urlretrieve(url, "./python.png")
# print(header1)
```

```python
import urllib.request as req

url = "http://encar.com/"

response=req.urlopen(url)
print(response)

print(f"url {response.geturl()}")
print(f"url {response.status}")
print(f"url {response.getheaders()}")

# html 코드 가져오기
contents = response.read().decode("euc-kr")  # 인고팅한게 EUC-KR 이라 디코딩도 이걸로 검
print(contents[:4000])
```

### requests
- pip install requests
- ullib.request 모듈과 같은 역할
- 디코딩도 알아서 해주고, json 처리 편함
- get, post, put, delete 메소드 제공함
- 세션 적용 가능


```python
# pip intstall requests
# pip install fake_useragent

import requests

url = "https://encar.com/"

response = requests.get(url)
print(response.status_code)
print(response.headers['content-type'])
print(response.text)
```

```python
# 세션 이용

session = requests.Session()

response = session.get(url)
print(response.status_code)
print(response.headers['content-type'])
print(response.text)

session.close()
```

```python
# get + 파라메터
# ?page=1&size=10
url = "https://httpbin.org/get"

params = {"test": "name"}

res= requests.get(url, params=params)
print(res.headers)
print(res.text)
print(res.json)
```

```python
# post + form data

data = {"name":"hong"}

res = requests.post("https://httpbin.org/post", data= data)
print(res.text)
```

```python
url = "https://jsonplaceholder.typicode.com/todos/1"

res = requests.get(url)
print(f"headers {res.headers}")  # 헤더 정보 출력
print(f"content {res.json()}")  # 헤더 정보 출력

contents = res.json()
print(contents.keys())
print(contents.values())
```

```python
url = "https://jsonplaceholder.typicode.com/users"

res = requests.get(url)
contents = res.json()

for line in contents:
    for k, v in line.items():
        print("Key : {}, Value : {}" .format(k, v) )
    print()
```

```python
# 행정안전부 rss
#https://www.mois.go.kr/gpms/view/jsp/rss/rss.jsp?ctxCd=1012
#https://www.mois.go.kr/gpms/view/jsp/rss/rss.jsp?ctxCd=1013
#https://www.mois.go.kr/gpms/view/jsp/rss/rss.jsp?ctxCd=1001

params = []
for code in [1012, 1013, 101]:
    params.append(dict(ctxCd=code))

print(params)
url = "https://www.mois.go.kr/gpms/view/jsp/rss/rss.jsp"

with requests.Session() as s:
    for param in params:
        res = s.get(url, params=param)
        print("-"*100)
        print(res.text)
```

### BeautifulSoup

```python
import requests
from bs4 import BeautifulSoup


url = "https://n.news.naver.com/mnews/article/021/0002715582"
res = requests.get(url)

# BeautifulSoup(내용, 파서지정)
soup = BeautifulSoup(res.text, "html.parser")
# print(soup.prettify())

# 원하는 요소 추출
# 1. 테그명 이용
head = soup.head
# print(head)
# print(head.text)

h2 = soup.h2
print(h2)
print(h2.text)
print(h2.attrs)
```
